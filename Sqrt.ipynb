```python
import numpy as np
import keras
from keras.layers import Dense,Activation
from keras.models import Sequential
import matplotlib.pyplot as plt
```

```python
number=np.array([2,3,4,5,6,7,8,9],dtype=int)
sqrt=np.array([4,9,16,25,36,49,64,81],dtype=int)
```

```python
for x,y in zip(number,sqrt):
    print(f"{x}--> {y} ")
    
```

```python
model=Sequential()   
model.add(Dense(units=1,activation='relu',input_shape=[1]))
model.add(Dense(units=1,activation='relu'))
```

```python
opt=keras.optimizers.Adam
model.summary()
model.compile(optimizer=opt(0.2),loss='mean_squared_error',metrics=['accuracy'])
model_last=model.fit(number,sqrt,epochs=1000)
```

```python
plt.plot(model_last.history["loss"])
```

```python
print(model.predict([9]))
```
